```python
#Importando bibliotecas & limpando os dados

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

url = "BTC_USD Bitfinex Historical Data.csv"
df = pd.read_csv(url, delimiter = ';', header = None, names = ["Data", "Close", "Open", "High", "Low"])
df = df.iloc[1:]
df = pd.DataFrame(df)

#ALTA (1) - BAIXA (0)
df.loc[(df.Close > df.Close.shift(1)), 'Alta/Baixa'] = 1
df.loc[(df.Close <= df.Close.shift(1)), 'Alta/Baixa'] = 0
df = df.fillna(0)

#LN (Variação Close atual/Close anterior)
df['Close'] = df['Close'].astype(float)
df['LN'] = (df['Close']/df['Close'].shift(1))-1
df = df.fillna(0)

#Média móvel(15 periodos) de LN
df['MEDIA'] = df['LN'].rolling(15).mean()
df = df.fillna(0)

#Desvio padrão (15 periodos) de LN
df.loc[(df['Alta/Baixa'] == 1), 'DESVIO'] = df['LN'].rolling(15).std()
df.loc[(df['Alta/Baixa'] == 0), 'DESVIO'] = -df['LN'].rolling(15).std()
df = df.fillna(0)

#Volatilidade acumulada (Soma dos últimos 15 valores do DESVIO)
df['VOLATILIDADE'] = df['DESVIO'].rolling(15).sum()
df = df.fillna(0)

df
```
